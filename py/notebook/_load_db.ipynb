```python
import MySQLdb as mdb
from pprint import pprint
from datetime import datetime
```

```python
db_host = '127.0.0.1'
db_user = 'root'
db_password = 'qwerty'
db_name = 'entity_determiner'
db_charset = 'utf8'
```

```python
tables = ['indonesia','dunia','bisnis','teknologi','hiburan','olahraga','science','kesehatan']

all_result = []
for t in tables:
    query = """SELECT * FROM {table_name} ORDER BY published_at DESC LIMIT 3""".format(table_name=t)
    
    connect = mdb.connect(db_host, db_user, db_password, db_name, charset=db_charset)
    cursor = connect.cursor()
    cursor.execute(query)
    result = cursor.fetchall()
    connect.close()
    
    all_res = []
    for r in result:
        json_result = {
            'id': r[0],
            'title': r[1],
            'clean_content': r[2][:50],
            'tagged_content': r[3][:50],
            'spoiler_content': r[4][:50],
            'url': r[5],
            'host': r[6],
            'published_at': r[7].strftime('%Y-%m-%d %H:%M:%S')
        }
        all_res.append(json_result)
    
    final_result = {
        'category': t,
        'items': all_res
    }
    all_result.append(final_result)
```
